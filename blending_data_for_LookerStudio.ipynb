```python
import pandas as pd
```

```python
orders = pd.read_csv('coffee_shop - clean_orders.csv')
products = pd.read_csv('coffee_shop - products.csv')
```

```python
# Cleaning Function (Removing '€' and converts to Float)
def clean_currency(x):
  if isinstance(x,str):
    return float(x.replace('€','').replace(',','.').strip())
  return x
```

```python
# Apply cleaning function to the releveant columns
orders['total_sale'] = orders['total_sale'].apply(clean_currency)
orders['unit_price'] = orders['unit_price'].apply(clean_currency)
products['profit'] = products['profit'].apply(clean_currency)
```

```python
# Merging Orders with Products
master_df = orders.merge(products, on='product_reference_number', suffixes=('_ord','_prod'))
```

```python
# Calculate total Profit per Order (Quanity * Unit Profit)
master_df['total_profit'] = master_df['quantity'] * master_df['profit']
```

```python
# Saving and Download the Master File
master_df.to_csv('coffee_shop_master_clean.csv', index=False)
print('Filed Saved! Downloaded coffee_shop_master_clean.csv')
```
